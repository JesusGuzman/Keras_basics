# Modelo VGG-16

```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
from keras.models import Model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Cargamos la imagen a clasificar

```python
img_path = './indice.jpeg'

img = image.load_img(img_path, target_size=(224, 224))

plt.imshow(img)
plt.show()
```

Keras nos provee de algunos modelos listos para usarse

```python
model = VGG16(include_top=True, weights='imagenet')
print(type(model))
```

```python
x = image.img_to_array(img)

x = np.expand_dims(x, axis=0)

x = preprocess_input(x)

print(x.max())

scores = model.predict(x)
```

Verificamos la clasificación

```python
class_table = open('./synset_words.txt', 'r')
lines = class_table.readlines()
print("scores type: ", type(scores))
print("scores shape: ", scores.shape)
print(np.argmax(scores))
print('result is ', lines[np.argmax(scores)])
class_table.close()
```
